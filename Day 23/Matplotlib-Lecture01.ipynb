## Create Labels for a Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([8, 12, 16, 20, 24, 28, 32, 36, 40, 44])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y,'s--g',lw='15',mec='b',mfc='y')

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

## Create a Title for a Plot
###### With Pyplot, you can use the title() function to set a title for the plot

```python

x = np.array([8, 12, 16, 20, 24, 28, 32, 36, 40, 44])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y,'s--g',lw='15',mec='b',mfc='y')

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Y axis VS X axis",loc='left')
plt.show()
```

## Matplotlib Adding Grid Lines
 grid() function to add grid lines to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([8, 12, 16, 20, 24, 28, 32, 36, 40, 44])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(axis = 'x')
# Parallel to y axis
plt.show()
```

### Matplotlib Subplots
###### subplots() function you can draw multiple plots in one figure
The subplots() function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are represented by the first and second argument.
The third argument represents the index of the current plot.
plt.subplot(i,j,1)
###### the figure has i row, j columns, and this plot is the first plot.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 3, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([1, 2, 3, 4])

plt.subplot(1, 3, 3)
plt.plot(x,y)

plt.show()
```

#### Super Title
##### You can add a title to the entire figure with the suptitle() function:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
y = np.array([30, 41, 32, 43])
x = np.array([1,2,3,4])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.xlabel("Subject No.")
plt.ylabel("Marks Obtained")
plt.title("3RD SEM")

#plot 2:
y = np.array([40, 41, 49, 46])
x = np.array([1, 2, 3, 4])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.xlabel("Subject No.")
plt.ylabel("Marks Obtained")
plt.title("4TH SEM")

plt.suptitle("MY MARKS")
plt.show()
```

### ColorMap
The Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.



This colormap is called 'viridis' and as you can see it ranges from 0, which is a purple color, and up to 100, which is a yellow color. Specifying the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.



```python
import matplotlib.pyplot as plt
import numpy as np



x = np.array([15,17,18,17,12,17,22,21,40,11,12,42,56])
y = np.array([29,46,77,38,11,36,63,81,44,68,73,82,89])
color=np.array([9,46,77,0,11,36,63,100,44,68,73,82,59])


plt.scatter(x, y,c=color, cmap='flag')

plt.show()
```

### Alpha
You can adjust the transparency of the dots with the alpha argument.
You can combine a colormap with different sizes on the dots.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.1, cmap='winter')

plt.colorbar()

plt.show()
```

### Matplotlib Histograms

Histogram
A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

In Matplotlib, we use the hist() function to create histograms.

The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

For simplicity we use NumPy to randomly generate an array with 150 values, where the values will concentrate around 70, and the standard deviation is 5.

```python
import numpy as np

x = np.random.normal(70, 5, 150)

print(x)
plt.hist(x,color='g',lw='0.5')
plt.show() 
```

#### Matplotlib Pie Charts

pie() function to draw pie charts

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

The size of each wedge is determined by comparing the value with all the other values, by using this formula:

The value divided by the sum of all values: x/sum(x)

### Adding Labels
##### The label parameter must be an array with one label for each wedge.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([40, 30, 15, 15])
mylabels = ["Audi", "BWM", "Mercedes", "Ferrari"]

plt.pie(y, labels = mylabels)
plt.show() 
```

Default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Audi", "BWM", "Mercedes", "Ferrari"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

### Explode
Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.
###### Adding a shadow to the pie chart by setting the shadows parameter to True.
### Legend
To add a list of explanation for each wedge, use the legend() function

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Mercedes", "BMW", "Audi", "Ferrari"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode,shadow = True)
plt.legend()
#Adding title to the legend
plt.legend(title = "Cars")
plt.show() 
```
