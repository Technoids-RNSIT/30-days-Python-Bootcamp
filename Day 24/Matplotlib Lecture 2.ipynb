### Stack Plot
Stackplot is used to draw a stacked area plot. It displays the complete data for visualization. It shows each part stacked onto one another and how each part makes the complete figure. It displays various constituents of data and it behaves like a pie chart. It has x-label, y-label and title in which various parts can be represented by different colors.

###### The idea of stack plots is to show “parts to the whole” over time. It is used to represent various datasets without overlapping over each other.



```python
import matplotlib.pyplot as plt
import numpy  as np

days = np.array([1, 2, 3, 4, 5])

Sleeping = np.array([7, 8, 6, 11, 7])
Studying = np.array([2 ,3 ,6 ,3 ,2])
playing =  np.array([8, 5, 7, 8, 13])

#baseline='wiggle'
#baseline='zero'
#baseline='sym'
plt.stackplot(days, Sleeping, Studying, playing,baseline='sym',colors =['hotpink', 'cyan','red'])
plt.legend(['Sleeping','Studying','playing'])
plt.xlabel('Days')

plt.ylabel('No of Hours')

plt.title('Representation of Studying Sleeping and Playing wrt to Days')

plt.show()
```

### Box Plot
##### A standardized way of displaying the distribution of data based on a five summary.
###### A box plot is also known as a whisker plot.
Displays a summary of minimum, first quartile(Q1), median(Q2),third quartile (Q3) and maximum.



```python
#[10,50,20,30,40]
#[10,20,30,40,50]
```

```python
import numpy as np
x=[10,50,20,30,40]
print(np.min(x))
print(np.quantile(x,[0.25,0.50,0.75]))
print(np.max(x))
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[10,50,20,30,40]
#labels=['M1']
#widths=0.8
#notch=True
#vert=False
plt.boxplot(x,labels=['Marks1'],widths=0.8,notch=True,vert=False)
plt.xlabel("Name of student")
plt.ylabel("Marks of student")
plt.title("Student Marks Statistics")
plt.show()
#Min=10,Max=50,Q1=20,Q2=30,Q3=40

```

### Matplotlib Heatmaps
A 2-D Heatmap is a data visualization tool that helps to represent the magnitude of the phenomenon in form of colors. In python, we can plot 2-D Heatmaps using Matplotlib package.
#### Matplotlib does not have a dedicated function for heatmap but we can build them using matplotlib’s imshow function.

```python
import numpy as np
import matplotlib.pyplot as plt
  
data = np.random.random(( 12 , 12 ))*100
print(data)
plt.imshow( data , cmap = 'viridis')
  
plt.title( "2-D Heat Map")
plt.show()
```

## Intro to Seaborn
Used for data visualization and based on Matplotlib

Seaborn allows the creation of statistical graphics

### Why Seaborn?
#### Seaborn fixes two shortcomings of Matplotlib:
- Matplotlib setting are difficult to figure out,Seaborn comes with numerous customized themes and high-level interfaces.
- Matplot doesn't serve well when it comes to dealing with DataFrames, while Seaborn functions actually work on DataFrames.
    

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset('fmri')
fmri.head()
#fmri.info()
```

```python
sns.lineplot(x='timepoint', y='signal', data = fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', data = fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', data = fmri,hue='event',style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', data = fmri,hue='event',style='event',marker='o')
plt.show()
```

### Distribution plot
It is used to draw a plot of two variables with bivariate and univariate graphs.

```python
df=pd.read_csv("diamonds.csv")
df.head()
#hist=False
#color='green'
#kde=False
# bins=50
#vertical=True
sns.distplot(df['price'],bins=50,hist=False,vertical=True,color='hotpink')
plt.show()
```
